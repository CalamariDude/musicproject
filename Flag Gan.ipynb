## We will be exploring a generative model to create fake flags

```python
import numpy as np
import pandas as pd
import cv2 as cv
import matplotlib.pyplot as plt
import os
%matplotlib inline

flags = []
path = "./datasets/country-flags/png100px/"
shape = (6,4)
for item in os.listdir(path):
    data = cv.imread(path + item)
    rescaled = cv.resize(data, shape)
    flags.append(rescaled)
    
print("Data shape = ", np.asarray(flags).shape)
```

## First let's see what our data looks like:

```python
plt.figure(figsize=(100,100))
for i in range(30):
    plt.subplot(10,3,i+1)
    plt.imshow(flags[i])
    
```

## Now let's define our generative architecture
We will be using an autoencoder for now

```python
from keras.layers import Input, Dense 
from keras.models import Model


encoding_dim = 

input_img = Input(shape=(shape[0] * shape[1]*3,))
encoded = Dense(encoding_dim, activation='relu')(input_img)
decoded = decoded = Dense(shape[0] * shape[1]*3, activation='sigmoid')(encoded)
autoencoder = Model(input_img, decoded)

# this model maps an input to its encoded representation
encoder = Model(input_img, encoded)

# create a placeholder for an encoded (32-dimensional) input
encoded_input = Input(shape=(encoding_dim,))
# retrieve the last layer of the autoencoder model
decoder_layer = autoencoder.layers[-1]
# create the decoder model
decoder = Model(encoded_input, decoder_layer(encoded_input))

autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
flattened_flags = np.asarray(flags).reshape((256,shape[0] * shape[1]*3))
print(flattened_flags.shape)
train_data = flattened_flags[:1]
autoencoder.fit(train_data, train_data,
                epochs=500,
                batch_size=1,
                shuffle=True,
                validation_data=(train_data, train_data))
```

```python
i= 0
encoded_imgs = encoder.predict(flattened_flags[i].reshape(1,-1))
decoded_imgs = decoder.predict(encoded_imgs)

rescaled = decoded_imgs[0].reshape((shape[1], shape[0], 3))
plt.figure()
plt.imshow(flattened_flags[i].reshape((shape[1], shape[0], 3)))
plt.figure()
plt.imshow(rescaled)
```
